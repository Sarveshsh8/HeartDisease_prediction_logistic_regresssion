Splitting the datasets and training the data

```python
X = df.drop(['TenYearCHD'], axis=1)
y = df['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100)
LogModel = LogisticRegression()
LogModel.fit(X_train, y_train)
```

```python
prediction = LogModel.predict(X_test)
```

```python
print(classification_report(y_test,prediction))
print(confusion_matrix(y_test,prediction))
```

```python
LogModel.score(X_train, y_train)
```

```python
LogModel.score(X_test, y_test)
```

USING CROSS VALIDATION TECHNIQUE

```python
CV = cross_val_score(LogisticRegression(), X, y, cv=10)
```

```python
print(f"After k-fold cross validation score is {CV.mean()}")
```

# GridSearch CV to get the best parameters for this model

```python
from sklearn.model_selection import GridSearchCV
param = [{'penalty' : ['l1', 'l2']},
            {'C': [1, 10, 100, 1000] }]

grid_search = GridSearchCV(estimator=LogModel,param_grid=param, scoring='accuracy', cv=5, verbose=0 )
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

CONCLUSION:
Model accuracy score for logistic regression after cross-validation is 84.87% which is quite nice.
